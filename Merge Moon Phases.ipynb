```python
import pandas as pd
import numpy
```

```python
moon_phases_2016 = "moon_phases_2016.csv"
moon_phases_2017 = "moon_phases_2017.csv"
moon_phases_2018 = "moon_phases_2018.csv"
```

```python
moon_phases_2016_df = pd.read_csv(moon_phases_2016)
moon_phases_2017_df = pd.read_csv(moon_phases_2017)
moon_phases_2018_df = pd.read_csv(moon_phases_2018)
```
